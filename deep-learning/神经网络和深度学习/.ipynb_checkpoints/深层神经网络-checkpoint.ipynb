## 前向传播和反向传播的实现

![060](images/060.png)

![061](images/061.png)

![062](images/062.png)

**循环以下几步：**

+ $$d_Z^{[l]}=d_A^{[l]}*g^{[l]}(Z^{[l]})    （1式）$$

+ $$d_W^{[l]}=\frac{1}{m}*d_Z^{[l]}{A^{[l-1]}}^T （2式）$$

+ $$d_b^{[l]}=\frac{1}{m}*np.sum(d_Z^{[l]},axis=1,keepdims=True) （3式）$$

+ $$d_A^{[l-1]}={W^{[l]}}^T*d_Z^{[l]} （4式）$$

依此循环进入一式。
